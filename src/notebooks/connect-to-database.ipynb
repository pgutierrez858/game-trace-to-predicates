```kotlin
val tableName = "tictactoe_game_over"

val df = DataFrame.readSqlTable(dbConfig, tableName, 5)

df
```

```kotlin
// Execute a SQL query
val df = DataFrame.readSqlQuery(dbConfig, """
    SELECT tictactoe_action_taken.GameID,
    tictactoe_action_taken.`BoardState(00X)`,
    tictactoe_action_taken.`BoardState(01X)`,
    tictactoe_action_taken.`BoardState(02X)`,
    tictactoe_action_taken.`BoardState(10X)`,
    tictactoe_action_taken.`BoardState(11X)`,
    tictactoe_action_taken.`BoardState(12X)`,
    tictactoe_action_taken.`BoardState(20X)`,
    tictactoe_action_taken.`BoardState(21X)`,
    tictactoe_action_taken.`BoardState(22X)`,
    tictactoe_action_taken.`BoardState(00O)`,
    tictactoe_action_taken.`BoardState(01O)`,
    tictactoe_action_taken.`BoardState(02O)`,
    tictactoe_action_taken.`BoardState(10O)`,
    tictactoe_action_taken.`BoardState(11O)`,
    tictactoe_action_taken.`BoardState(12O)`,
    tictactoe_action_taken.`BoardState(20O)`,
    tictactoe_action_taken.`BoardState(21O)`,
    tictactoe_action_taken.`BoardState(22O)`,
    tictactoe_game_over.`Winner(PlayerIdx)`
    FROM tictactoe_action_taken
    LEFT JOIN tictactoe_game_over ON tictactoe_action_taken.GameID = tictactoe_game_over.GameID
    WHERE tictactoe_game_over.`Winner(PlayerIdx)` != -1
""")

// Print the DataFrame
df
```

```kotlin
val renamedDf = df.rename(
    "BoardState(00X)" to "x00", "BoardState(01X)" to "x01", "BoardState(02X)" to "x02",
    "BoardState(10X)" to "x10", "BoardState(11X)" to "x11", "BoardState(12X)" to "x12",
    "BoardState(20X)" to "x20", "BoardState(21X)" to "x21", "BoardState(22X)" to "x22",
    "BoardState(00O)" to "o00", "BoardState(01O)" to "o01", "BoardState(02O)" to "o02",
    "BoardState(10O)" to "o10", "BoardState(11O)" to "o11", "BoardState(12O)" to "o12",
    "BoardState(20O)" to "o20", "BoardState(21O)" to "o21", "BoardState(22O)" to "o22",
    "Winner(PlayerIdx)" to "winner")
renamedDf
```

```kotlin
val grouppedGames = renamedDf.groupBy { GameID and winner }

grouppedGames
```

```kotlin
fun createFolder(folderPath: String): Boolean {
    val folder = File(folderPath)
    if (!folder.exists()) {
        return folder.mkdirs()
    }
    return true
}
val folderPath = "./../main/resources/samples/tictactoe/"
createFolder(folderPath)
createFolder(folderPath + "examples")
createFolder(folderPath + "counter-examples")

grouppedGames.forEach { gameGroup ->
    gameGroup.group.writeCSV(
        folderPath + (if (gameGroup.key.winner == 0) "examples" else "counter-examples") +
                "/${gameGroup.key.GameID}.csv"
    )
}
```
